### Exercise 1. 

Read Total Profit for all months and show it using a line plot with Month Number as your baseline

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
# Exercise 1

import pandas as pd


df_1 = pd.read_csv("company_sales_data.csv")
df_1.head()
````

````python
# Exercise 1
import matplotlib.pyplot as plt

X = df_1['month_number']
y = df_1['total_profit']

fig = plt.figure()
ax = plt.axes()

ax.plot(X,y)
ax.grid()
ax.set_ylabel('total profit')
ax.set_xlabel('month number')
ax.set_title('Total Profit of all Months')
````

### Exercise 2. 

Get Total Units Sold of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Units Sold Number
5. Add a circle marker.

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

### Exercise 6.
Write a script that creates your favorite visual type you have learned about, and display the visual once from within the notebook:

````python
# src: https://realpython.com/mandelbrot-set-python/

def complex_matrix(xmin, xmax, ymin, ymax, pixel_density):
    re = np.linspace(xmin, xmax, int((xmax - xmin) * pixel_density))
    im = np.linspace(ymin, ymax, int((ymax - ymin) * pixel_density))
    return re[np.newaxis, :] + im[:, np.newaxis] * 1j

def is_stable(c, num_iterations):
    z = 0
    for _ in range(num_iterations):
        z = z ** 2 + c
    return abs(z) <= 2

def get_members(c, num_iterations):
    mask = is_stable(c, num_iterations)
    return c[mask]

import matplotlib.pyplot as plt
import numpy as np

np.warnings.filterwarnings("ignore")

c = complex_matrix(-2, 0.5, -1.5, 1.5, pixel_density=512)
plt.imshow(is_stable(c, num_iterations=20), cmap="binary")
plt.gca().set_aspect("equal")
plt.axis("off")
plt.tight_layout()
````
